```python
import matplotlib as mpl
mpl.use('TkAgg')
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
from matplotlib import colors
import pandas as pd
import numpy as np

font = FontProperties()
font.set_name('Arial')
plt.rcParams['axes.unicode_minus']=False
plt.rcParams['font.family']='Arial'

file_path = r'./angleData.xlsx'
s_x = pd.read_excel(file_path, sheet_name = "O1-Te-O4", usecols=[0])
s_arm = pd.read_excel(file_path, sheet_name = "O1-Te-O4", usecols=[1]) 
s_zig = pd.read_excel(file_path, sheet_name = "O1-Te-O4", usecols=[2]) 

array_x = np.array(list(map(float, np.array(s_x))))
array_arm = np.array(list(map(float, np.array(s_arm))))
array_zig = np.array(list(map(float, np.array(s_zig))))

linear_strain1 = np.polyfit(array_x,array_arm,1)
linear_strain2 = np.polyfit(array_x,array_zig,1)

l = plt.plot(array_x,linear_strain1[0]*array_x+linear_strain1[1],c=(134/255, 147/255, 117/255),linestyle='--',lw=2)
l = plt.plot(array_x,linear_strain2[0]*array_x+linear_strain2[1],c=(109/255, 154/255, 163/255),linestyle='--',lw=2)

s = plt.scatter(array_x,array_arm,color=(134/255, 147/255, 117/255), marker='o',label='Armchair', s=60)
s = plt.scatter(array_x,array_zig,color=(109/255, 154/255, 163/255), marker='o',label='Zigzag', s=60)

plt.xlim([-1.25,1.25]) # x range
plt.ylim([74.5,75.25]) # y range
plt.ylabel('$∠$Te1-O-Te4 (°)',fontsize=20, fontproperties=font)
plt.xlabel('Strain (%)',fontsize=20, fontproperties=font)
plt.xticks(size = 16)
plt.yticks(size = 16)
plt.legend(scatterpoints =1, numpoints=1,markerscale=1.0, bbox_to_anchor=(0.65, 0.225), loc='upper left', borderaxespad=0.5, frameon=False, handletextpad=0.1, fontsize=16)   # The location of legend.

plt.savefig('O1-Te-O4.png', bbox_inches = 'tight', dpi=500, transparent=True)
plt.show()
plt.close()
```

```python
import matplotlib as mpl
mpl.use('TkAgg')
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
from matplotlib import colors
import pandas as pd
import numpy as np

font = FontProperties()
font.set_name('Arial')
plt.rcParams['axes.unicode_minus']=False

file_path = r'./angleData.xlsx'
s_x = pd.read_excel(file_path, sheet_name = "Te-O2-Te", usecols=[0])
s_arm = pd.read_excel(file_path, sheet_name = "Te-O2-Te", usecols=[1]) 
s_zig = pd.read_excel(file_path, sheet_name = "Te-O2-Te", usecols=[2]) 

array_x = np.array(list(map(float, np.array(s_x))))
array_arm = np.array(list(map(float, np.array(s_arm))))
array_zig = np.array(list(map(float, np.array(s_zig))))

s = plt.plot(array_x,array_arm,color='darkorange', linestyle='--',marker='*',markeredgecolor='darkorange',label='Armchair')
s = plt.plot(array_x,array_zig,color='forestgreen', linestyle='--',marker='o',markeredgecolor='forestgreen',label='Zigzag')

plt.xlim([-1.25,1.25]) # x range
plt.ylim([136.5,138]) # y range
plt.ylabel('$∠$Te-O2-Te (°)',fontsize=20, fontproperties=font)
plt.xlabel('Strain (%)',fontsize=20, fontproperties=font)
plt.xticks(size = 16)
plt.yticks(size = 16)
plt.legend(scatterpoints =1, numpoints=1,markerscale=1.0, bbox_to_anchor=(0.7, 0.225), loc='upper left', borderaxespad=0.5, frameon=False, handletextpad=0.1, fontsize=16)   # The location of legend.

plt.savefig('Te-O2-Te.png', bbox_inches = 'tight', dpi=500, transparent=True)
plt.show()
plt.close()
```
