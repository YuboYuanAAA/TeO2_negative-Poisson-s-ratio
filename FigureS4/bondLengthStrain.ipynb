```python
import matplotlib as mpl
mpl.use('TkAgg')
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
from matplotlib import colors
import pandas as pd
import numpy as np

font = FontProperties()
font.set_name('Arial')
plt.rcParams['axes.unicode_minus']=False
plt.rcParams['font.family']='Arial'

file_path = r'./bondlengthData.xlsx'
s_x = pd.read_excel(file_path, sheet_name = "Armchair", usecols=[0])
s_To1 = pd.read_excel(file_path, sheet_name = "Armchair", usecols=[11]) 
s_To2 = pd.read_excel(file_path, sheet_name = "Armchair", usecols=[12]) 
s_To3 = pd.read_excel(file_path, sheet_name = "Armchair", usecols=[13]) 
s_To4 = pd.read_excel(file_path, sheet_name = "Armchair", usecols=[14]) 
s_ave = pd.read_excel(file_path, sheet_name = "Armchair", usecols=[15]) 

array_x = np.array(list(map(float, np.array(s_x))))
array_To1 = np.array(list(map(float, np.array(s_To1))))
array_To2 = np.array(list(map(float, np.array(s_To2))))
array_To3 = np.array(list(map(float, np.array(s_To3))))
array_To4 = np.array(list(map(float, np.array(s_To4))))
array_ave = np.array(list(map(float, np.array(s_ave))))

linear_strain1 = np.polyfit(array_x,array_To1,1)
linear_strain2 = np.polyfit(array_x,array_To2,1)
linear_strain3 = np.polyfit(array_x,array_To3,1)
linear_strain4 = np.polyfit(array_x,array_To4,1)
linear_strainAve = np.polyfit(array_x,array_ave,1)

l = plt.plot(array_x,linear_strain1[0]*array_x+linear_strain1[1],c='lightskyblue',linestyle='--',lw=2)
l = plt.plot(array_x,linear_strain2[0]*array_x+linear_strain2[1],c='deepskyblue',linestyle='--',lw=2)
l = plt.plot(array_x,linear_strain3[0]*array_x+linear_strain3[1],c='lightcoral',linestyle='--',lw=2)
l = plt.plot(array_x,linear_strain4[0]*array_x+linear_strain4[1],c='crimson',linestyle='--',lw=2)
l = plt.plot(array_x,linear_strainAve[0]*array_x+linear_strainAve[1],c='grey',linestyle='--',lw=2)

s = plt.scatter(array_x,array_To1,color='lightskyblue', marker='o',label='Te-O1')
s = plt.scatter(array_x,array_To2,color='dodgerblue', marker='o',label='Te-O2')
s = plt.scatter(array_x,array_To3,color='lightcoral', marker='o',label='Te-O3')
s = plt.scatter(array_x,array_To4,color='crimson', marker='o',label='Te-O4')
s = plt.scatter(array_x,array_ave,color='grey', marker='*',label='Average',s=70)

plt.xlim([-1.25,1.25]) # x range
plt.ylim([-0.2,0.2]) # y range
plt.ylabel('Bond length (%)',fontsize=20, fontproperties=font)
plt.xlabel('Amchair strain (%)',fontsize=20, fontproperties=font)
plt.xticks(size = 16)
plt.yticks(size = 16)
plt.legend(scatterpoints =1, numpoints=1,markerscale=1.0, bbox_to_anchor=(0.3, 0.50), loc='upper left', borderaxespad=0.5, frameon=False, handletextpad=0.1, fontsize=16)   # The location of legend.

plt.savefig('Armchair.png', bbox_inches = 'tight', dpi=500, transparent=True)
plt.show()
plt.close()
```
